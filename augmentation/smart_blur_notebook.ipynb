```python


# Make .lab files next to each WAV for MFA (Option A)
# - Leaves original timed .txt transcripts untouched
# - Writes one-line plain-text transcripts as <basename>.lab beside the WAV

from pathlib import Path
import re

# ====== EDIT ROOT ONLY ======
DATASET_ROOT = Path(r"E:\TCD_TIMIT")   # <-- the folder that contains /volunteers and /lipspeakers

# ===========================
# Use the same toggle from Cell 2
for spk in SPEAKERS_TO_PROCESS:
    VOL_ROOT = DATASET_ROOT / SPEAKER_GROUP_NAME / spk / "Clips"

    AUDIO_ROOTS = [VOL_ROOT / "processed" / "audio16k"]   # searches all subfolders (e.g., straightcam/)
    TRANSCRIPT_DIRS_TRY = [
        VOL_ROOT / "transcripts" / CAMERA,   # prefer matching camera
        VOL_ROOT / "transcripts" / "straightcam",  # fallback (if your transcripts are only here)
        VOL_ROOT / "transcripts",            # last fallback
    ]

    def normalize_word(w: str) -> str:
        """
        Light normalization for MFA:
        - uppercase (matches many CMU/MFA dict conventions)
        - keep letters and apostrophes
        - drop everything else
        """
        w = w.strip()
        w = w.replace("â€™", "'").replace("â€˜", "'")
        w = re.sub(r"[^A-Za-z']+", "", w)
        return w.upper()

    def find_timed_txt(stem_lower: str):
        """Return the first transcript Path whose stem matches (case-insensitive), else None."""
        for tdir in TRANSCRIPT_DIRS_TRY:
            if not tdir.exists():
                continue
            for p in tdir.glob("*.txt"):
                if p.stem.lower() == stem_lower:
                    return p
        return None

    def words_from_timed_txt(p: Path):
        """
        Parse either:
          - timed lines like: '2940 5200 she' -> take the 3rd token
          - or plain text line(s) -> split into tokens
        """
        words = []
        lines = p.read_text(encoding="utf-8", errors="ignore").splitlines()

        for line in lines:
            parts = line.strip().split()
            if len(parts) >= 3 and parts[0].isdigit() and parts[1].isdigit():
                w = normalize_word(parts[2])
                if w:
                    words.append(w)
            elif line.strip():
                # treat as plain text transcript
                for tok in line.strip().split():
                    w = normalize_word(tok)
                    if w:
                        words.append(w)
                # if it's plain text, we can stop after first non-empty line
                break

        return words

    # ---- sanity checks ----
    if not VOL_ROOT.exists():
        raise FileNotFoundError(f"VOL_ROOT does not exist: {VOL_ROOT}")

    made, skipped = 0, 0
    for ar in AUDIO_ROOTS:
        if not ar.exists():
            raise FileNotFoundError(f"Missing audio root: {ar}")

        for wav in ar.rglob("*.wav"):
            stem_lower = wav.stem.lower()  # e.g., 'sa1'
            txt = find_timed_txt(stem_lower)

            if txt is None:
                print(f"[SKIP] No transcript for {wav.name}")
                skipped += 1
                continue

            words = words_from_timed_txt(txt)
            if not words:
                print(f"[WARN] Empty/invalid transcript in {txt}")
                skipped += 1
                continue

            lab_path = wav.with_suffix(".lab")
            lab_path.write_text(" ".join(words) + "\n", encoding="utf-8")
            print(f"[OK] {lab_path}  <-- from  {txt.name}")
            made += 1

    print(f"\nDone. Wrote {made} .lab files; skipped {skipped}.")
    print(f"MFA tip: run MFA on {AUDIO_ROOTS[0]} (or a camera subfolder). It will pick up .wav + .lab pairs.")
```

### dictionary of viseme to phoneme mapping


```python
# Phone to Viseme mapping aligned to MFA and TIMIT English

import re

# Viseme families used elsewhere: "/A"..."/K", "/S"
# We map normalized phones (mostly IPA, some ARPAbet bridged) to those families.

VISEME_MAP = {
    # /A Labiodentals
    "f": "/A", "v": "/A",

    # /B Rounded-back vowels and rhotics/glides (puckered/rounded)
    "u": "/B", "o": "/B", "É”": "/B", "ÊŠ": "/B", "É’": "/B", "Éµ": "/B", "Ê‰": "/B", "y": "/B",
    "É¹": "/B", "r": "/B", "Éš": "/B", "É": "/B", "w": "/B",
    "ow": "/B", "oÊŠ": "/B", "É™w": "/B",  # MFA sometimes outputs ow/oÊŠ/É™w

    # /C Bilabials
    "b": "/C", "p": "/C", "m": "/C",

    # /D Diphthong /aw/
    "aÊŠ": "/D",

    # /E Interdentals (true tongue-between-teeth)
    "Ã°": "/E", "Î¸": "/E",

    # /F Postalveolar sibilants and affricates
    "Êƒ": "/F", "Ê’": "/F", "tÊƒ": "/F", "dÊ’": "/F",

    # /G Mid-back rounded (/oy/ /ao/)
    "É”Éª": "/G", "É”j": "/G", "ao": "/G",

    # /H Alveolar sibilants
    "s": "/H", "z": "/H",

    # /I Unrounded/front-central vowels and neutral diphthongs
    "a": "/I", "Ã¦": "/I", "É›": "/I", "e": "/I", "i": "/I", "Éª": "/I",
    "É™": "/I", "Éœ": "/I", "É": "/I", "ÊŒ": "/I", "É‘": "/I",
    "aÉª": "/I", "eÉª": "/I",
    "aj": "/I", "ej": "/I",  # MFA variants of aÉª/eÉª

    # /J Alveolar non-sibilants (stops, nasals, laterals)
    "t": "/J", "d": "/J", "tÌª": "/J", "dÌª": "/J",  # include TH-stopping dental stops
    "n": "/J", "l": "/J", "É«": "/J", "É¾": "/J",

    # /K Velars / back-of-mouth stops (plus common MFA extras)
    "g": "/K", "É¡": "/K", "k": "/K", "Å‹": "/K",  # include script-g É¡
    "ÉŸ": "/K", "c": "/K", "kp": "/K", "É¡b": "/K",  # palatal/labiovelars seen in english_mfa

    # /S Silence/closures/non-speech (only if you want to treat them as a family)
    "h": "/S", "Ê”": "/S",
}

# Additional tokens treated as ignore (silence, closures, noise) during mapping:
_IGNORE = {
    "sp", "sil", "spn", "pau", "h#", "brth", "nsn",
    "pcl", "bcl", "tcl", "dcl", "kcl", "gcl", "q",   # closures + glottal stop (TIMIT)
}

# Keep dental/underring diacritics (tÌª/dÌª) because they are meaningful here.
# Most other diacritics can be stripped safely for viseme mapping.
_KEEP_COMBINING = {"\u032A", "\u033A"}  # COMBINING BRIDGE BELOW / COMBINING INVERTED BRIDGE BELOW

def _strip_combining(s: str) -> str:
    import unicodedata as ud
    s_nfd = ud.normalize("NFD", s)
    out = []
    for ch in s_nfd:
        if ud.category(ch) == "Mn" and ch not in _KEEP_COMBINING:
            continue
        out.append(ch)
    return ud.normalize("NFC", "".join(out))

def normalize_phone(ph: str) -> str:
    """
    Normalize a phone string from MFA/TIMIT into a canonical key we can map:
      - lowercase
      - strip ARPAbet stress digits (IY1 -> iy)
      - convert common ARPAbet to IPA
      - remove silence/closure tokens
      - collapse affricate split forms (t Êƒ -> tÊƒ)
      - remove length/rounding diacritics (but keep dental diacritics tÌª/dÌª)
    """
    ph = ph.strip().lower()
    if not ph:
        return ""

    # unify common TIMIT/MFA silence tokens
    if ph in _IGNORE:
        return ""

    # strip vowel stress digits (iy1 -> iy)
    ph = re.sub(r"([a-z]+)[0-2]$", r"\1", ph)

    # normalize special TIMIT variants
    ph = ph.replace("ax-h", "ax")  # reduced schwa before h
    ph = ph.replace("h#", "sil")

    # affricate joiners
    ph = ph.replace("tÍ¡Êƒ", "tÊƒ").replace("dÍ¡Ê’", "dÊ’")
    ph = re.sub(r"\bt\s*Êƒ\b", "tÊƒ", ph)
    ph = re.sub(r"\bd\s*Ê’\b", "dÊ’", ph)

    # common IPA modifier letters that are not useful for viseme buckets
    ph = ph.replace("Ë", "").replace("Ê°", "").replace("Ê·", "").replace("Ê²", "").replace("Ìƒ", "")

    # ARPAbet to IPA bridge (common English phones)
    ARPA2IPA = {
        # vowels/diphthongs
        "aa": "É‘", "ae": "Ã¦", "ah": "ÊŒ", "ao": "ao", "aw": "aÊŠ", "ax": "É™", "axr": "Éš",
        "ay": "aÉª", "eh": "É›", "er": "É", "ey": "eÉª", "ih": "Éª", "iy": "i",
        "ow": "oÊŠ", "oy": "É”Éª", "uh": "ÊŠ", "uw": "u", "ux": "Ê‰", "ix": "Éª",
        # consonants
        "sh": "Êƒ", "zh": "Ê’", "ch": "tÊƒ", "jh": "dÊ’",
        "dh": "Ã°", "th": "Î¸",
        "ng": "Å‹", "hh": "h", "r": "É¹", "y": "j", "dx": "É¾",
        # sometimes seen in resources
        "q": "Ê”",
        # syllabic consonants (map to base)
        "em": "m", "en": "n", "eng": "Å‹", "el": "l", "nx": "n",
        # glides
        "w": "w",
    }
    ph = ARPA2IPA.get(ph, ph)

    ph = _strip_combining(ph)
    return ph

def phone_to_viseme(ph: str):
    ph = normalize_phone(ph)
    if not ph:
        return None

    # direct hit
    if ph in VISEME_MAP:
        return VISEME_MAP[ph]

    # direct fallbacks for common phones
    if ph in {"p","b","m"}: return "/C"
    if ph in {"f","v"}: return "/A"
    if ph in {"t","d","tÌª","dÌª","n","l","É¾"}: return "/J"
    if ph in {"s","z"}: return "/H"
    if ph in {"Êƒ","Ê’","tÊƒ","dÊ’"}: return "/F"
    if ph in {"k","g","É¡","Å‹","ÉŸ","c","kp","É¡b"}: return "/K"
    if ph in {"w","É¹","r"}: return "/B"
    if ph in {"h","Ê”"}: return "/S"

    # vowel fallback: unknown vowel-like tokens -> /I (spread/neutral)
    if any(ch in "aeiouÉªÊŠÉ›Ã¦É‘É’É”É™ÉœÉÊŒyÊ‰Éµ" for ch in ph):
        return "/I"

    return None
```

### Get active spans from TextGrid

```python
def words_with_viseme_spans(textgrid_path: Path, target_viseme: str, lead=0.0, lag=0.0):
    """
    (Legacy helper, kept for reference)

    [(word, t0, t1)] for words whose phones include the target viseme.
    lead/lag expand the span for coarticulation coverage.
    """
    tg = tgt.io.read_textgrid(str(textgrid_path))
    words_tier  = tg.get_tier_by_name("words")
    phones_tier = tg.get_tier_by_name("phones")

    spans = []
    for w in words_tier.intervals:
        if not w.text:
            continue
        w0, w1 = w.start_time, w.end_time
        hit = False
        for p in phones_tier.intervals:
            if p.end_time <= w0 or p.start_time >= w1:
                continue
            v = phone_to_viseme(p.text)
            if v == target_viseme:
                hit = True
                break
        if hit:
            spans.append((w.text, max(0.0, w0 - lead), w1 + lag))
    return spans


def get_target_spans(textgrid_path: Path, target_visemes, span_mode: str, lead=0.0, lag=0.0):
    """
    Return a list of spans tagged with the viseme to use for blurring:

      spans = [(viseme, t0, t1), ...]

    span_mode:
      - "word" : one span per word that contains ANY target viseme.
                 If multiple target visemes occur in the word, we assign the span to the
                 *highest-priority* viseme based on the order in `target_visemes`.
      - "phone": one span per phone interval whose viseme is in target_visemes.
    """
    if isinstance(target_visemes, str):
        target_visemes = [target_visemes]
    assert span_mode in ("word", "phone")

    tg = tgt.io.read_textgrid(str(textgrid_path))
    words_tier  = tg.get_tier_by_name("words")
    phones_tier = tg.get_tier_by_name("phones")

    target_set = set(target_visemes)
    priority = {v: i for i, v in enumerate(target_visemes)}

    spans = []

    if span_mode == "phone":
        for p in phones_tier.intervals:
            if not p.text:
                continue
            v = phone_to_viseme(p.text)
            if v in target_set:
                spans.append((v, max(0.0, p.start_time - lead), p.end_time + lag))
        return spans

    # span_mode == "word"
    for w in words_tier.intervals:
        if not w.text:
            continue
        w0, w1 = w.start_time, w.end_time

        hits = []
        for p in phones_tier.intervals:
            if p.end_time <= w0 or p.start_time >= w1:
                continue
            v = phone_to_viseme(p.text)
            if v in target_set:
                hits.append(v)

        if hits:
            # choose the highest-priority target viseme found in this word
            chosen = min(hits, key=lambda vv: priority.get(vv, 10**9))
            spans.append((chosen, max(0.0, w0 - lead), w1 + lag))

    return spans


def spans_to_frames_viseme(spans, fps: float):
    """
    [(viseme, t0, t1)] â†’ [(viseme, f0, f1)] (inclusive)
    """
    out = []
    for viseme, t0, t1 in spans:
        f0 = int(math.floor(t0 * fps))
        f1 = int(math.ceil(t1 * fps)) - 1
        out.append((viseme, f0, f1))
    return out


# quick peek (optional) - won't crash if TEXTGRID isn't defined in your session
if "TEXTGRID" in globals():
    vis_spans = get_target_spans(TEXTGRID, TARGET_VISEMES, span_mode=SPAN_MODE, lead=LEAD_SEC, lag=LAG_SEC)
    print(f"viseme spans (sec) sample [mode={SPAN_MODE}]:", vis_spans[:6])
else:
    print("NOTE: TEXTGRID not set (this is fine for batch mode).")
```

### Landmark loader & preview

```python
# dlib-68 mouth indices
MOUTH_OUTER = np.arange(48, 60)  # 48..59
MOUTH_INNER = np.arange(60, 68)  # 60..67

# region coloring for preview
# region coloring for preview
REGIONS = {
    "corner_left":  [48, 60],     # add inner-corner
    "corner_right": [54, 64],     # add inner-corner
    "upper_outer":  list(range(50, 53)),  # 50,51,52
    "lower_outer":  list(range(56, 59)),  # 56,57,58
    "upper_inner":  list(range(61, 64)),  # 61,62,63   <-- fixed
    "lower_inner":  list(range(65, 68)),  # 65,66,67
    "outer_all":    list(range(48, 60)),
    "inner_all":    list(range(60, 68)),
}
REGION_PRIORITY = [
    "corner_left","corner_right",
    "upper_outer","lower_outer",
    "upper_inner","lower_inner",
    "outer_all","inner_all"
]
COLOR_BY_REGION = {
    "corner_left":  (0, 0, 255),
    "corner_right": (255, 0, 0),
    "upper_outer":  (0, 255, 0),
    "lower_outer":  (0, 255, 255),
    "upper_inner":  (255, 0, 255),
    "lower_inner":  (255, 255, 0),
    "outer_all":    (160, 160, 160),
    "inner_all":    (200, 200, 200),
}

def load_landmarks_pkl(pkl_path: Path) -> np.ndarray:
    with open(pkl_path, "rb") as f:
        obj = pickle.load(f)

    def as_array(x):
        x = np.asarray(x)
        if x.shape == (68,2):
            return x
        raise ValueError(f"Frame not shaped (68,2), got {x.shape}")

    frames = []
    if isinstance(obj, list):
        for item in obj:
            frames.append(as_array(item["landmarks"] if isinstance(item, dict) and "landmarks" in item else item))
    elif isinstance(obj, dict):
        if "landmarks" in obj:
            for item in obj["landmarks"]:
                frames.append(as_array(item))
        elif "frames" in obj:
            for item in obj["frames"]:
                frames.append(as_array(item["landmarks"] if isinstance(item, dict) and "landmarks" in item else item))
        else:
            try: keys = sorted(obj.keys(), key=lambda k: int(k))
            except Exception: keys = list(obj.keys())
            for k in keys:
                item = obj[k]
                frames.append(as_array(item["landmarks"] if isinstance(item, dict) and "landmarks" in item else item))
    else:
        raise TypeError(f"Unsupported pkl root type: {type(obj)}")

    arr = np.stack(frames, 0)
    if arr.shape[1:] != (68,2):
        raise ValueError(f"Final shape not [T,68,2], got {arr.shape}")
    return arr

def _point_region(idx: int) -> str:
    for region in REGION_PRIORITY:
        if idx in REGIONS[region]:
            return region
    return "unassigned"

def draw_landmarks(img_bgr: np.ndarray, lm: np.ndarray, radius: int = 2) -> np.ndarray:
    out = img_bgr.copy()
    for i in range(48, 68):
        x, y = lm[i].astype(int)
        color = COLOR_BY_REGION.get(_point_region(i), (255,255,255))
        cv2.circle(out, (int(x), int(y)), radius, color, thickness=-1, lineType=cv2.LINE_AA)
        cv2.putText(out, str(i), (int(x)+3, int(y)-3), cv2.FONT_HERSHEY_SIMPLEX, 0.35, color, 1, cv2.LINE_AA)
    return out

# optional: preview image & CSV export
def single_frame_preview(pkl_path: Path, video_path: Path, frame_idx: int, out_png: Path):
    lms = load_landmarks_pkl(pkl_path)
    T = lms.shape[0]
    idx = max(0, min(frame_idx, T-1))
    if video_path.exists():
        cap = cv2.VideoCapture(str(video_path))
        cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
        ret, frame = cap.read()
        cap.release()
        if not ret:
            raise RuntimeError("Could not read preview frame")
        img = frame
    else:
        lm = lms[idx]
        minx, miny = lm[48:68,0].min(), lm[48:68,1].min()
        maxx, maxy = lm[48:68,0].max(), lm[48:68,1].max()
        pad = 40
        w = int((maxx - minx) + 2*pad)
        h = int((maxy - miny) + 2*pad)
        img = np.full((max(h,100), max(w,100), 3), 20, dtype=np.uint8)
        lms = lms.copy()
        lms[idx,:,0] = lms[idx,:,0] - minx + pad
        lms[idx,:,1] = lms[idx,:,1] - miny + pad

    annotated = draw_landmarks(img, lms[idx], radius=3)
    cv2.imwrite(str(out_png), annotated)
    print(f"Saved preview: {out_png}")

def export_csv(pkl_path: Path, out_csv: Path):
    lms = load_landmarks_pkl(pkl_path)
    with open(out_csv, "w", newline="") as f:
        writer = csv.writer(f)
        writer.writerow(["frame","landmark_idx","x","y","region"])
        for t in range(lms.shape[0]):
            for i in range(48,68):
                x, y = lms[t,i]
                writer.writerow([t, i, float(x), float(y), _point_region(i)])


```

### Visual Aspect - helper fucntions for blur

```python
# --- Blobs-only helpers (kept); legacy SIMPLE_MASKS & blur_lip_not_aoi REMOVED ---

def _soften(mask, blur=5):
    if blur < 1:
        return mask.astype(np.float32)
    m = cv2.GaussianBlur(mask*255, (2*blur+1, 2*blur+1), 0) / 255.0
    return m.astype(np.float32)

def _apply_local_blur(frame_bgr, roi_mask, ksize=201, sigma=0):
    """Blur ONLY inside roi_mask, keep the rest unchanged."""
    ksize = int(ksize) | 1  # odd
    roi = _soften(roi_mask, blur=3)[..., None]  # soft edges
    blurred = cv2.GaussianBlur(frame_bgr, (ksize, ksize), sigma)
    return (roi*blurred + (1.0-roi)*frame_bgr).astype(np.uint8)




```

```python
## --- Blobby ops we keep ---

def mask_quad_from_indices(shape, lm, idxs=(50, 52, 58, 56), pad=6):
    """
    Make a blob mask from the convex hull of the given landmark indices.
    pad dilates the blob so it looks indistinguishable.
    """
    H, W = shape[:2]
    m = np.zeros((H, W), dtype=np.uint8)
    pts = np.array([lm[i] for i in idxs], dtype=np.int32)
    hull = cv2.convexHull(pts)
    cv2.fillConvexPoly(m, hull, 1)
    if pad > 0:
        k = 2*int(pad)+1
        m = cv2.dilate(m, cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (k, k)))
    return m

def blur_specific_quad(frame_bgr, lm_68, idxs=(50, 52, 58, 56), pad=8, ksize=61):
    """
    Blur ONLY the quad region defined by idxs; everything else stays as-is.
    Use a large ksize + pad for an indistinguishable blob.
    """
    quad = mask_quad_from_indices(frame_bgr.shape, lm_68, idxs=idxs, pad=pad)
    return _apply_local_blur(frame_bgr, quad, ksize=int(ksize)|1, sigma=0)

def blur_inside_mouth_blob(frame_bgr, lm_68, pad=10, ksize=75):
    """
    Blur ONLY the intraoral quad (61,63,67,65). Everything else unchanged.
    """
    return blur_specific_quad(frame_bgr, lm_68, idxs=(61, 63, 67, 65), pad=pad, ksize=ksize)

def blur_lip_band_blob(frame_bgr, lm_68, which="upper", pad=10, ksize=75):
    """
    Blur ONLY a lip-band quad.
      which="upper" -> (50,52,61,63)  (upper central band)
      which="lower" -> (58,67,65,56)  (lower central band)
    """
    if which.lower() == "upper":
        idxs = (50, 52, 61, 63)
    elif which.lower() == "lower":
        idxs = (58, 67, 65, 56)
    else:
        raise ValueError("which must be 'upper' or 'lower'")
    return blur_specific_quad(frame_bgr, lm_68, idxs=idxs, pad=pad, ksize=ksize)

def blur_corners_blob(frame_bgr, lm_68, which="both", pad=10, ksize=75):
    """
    Blur the mouth corners using 'blobby' quads near each corner.
      left  quad â‰ˆ (48,49,60,59)
      right quad â‰ˆ (54,53,64,63)
    """
    which = which.lower()
    out = frame_bgr
    if which in ("left", "both"):
        out = blur_specific_quad(out, lm_68, idxs=(48, 49, 60, 59), pad=pad, ksize=ksize)
    if which in ("right", "both"):
        out = blur_specific_quad(out, lm_68, idxs=(54, 53, 64, 55), pad=pad, ksize=ksize)
    return out
```

```python
# Blobs to BLUR (non-interest) per viseme group.
# kinds: "corners", "lip_band" (which="upper"/"lower"), "inside_mouth", "quad"

VISEME_TO_BLOBS = {
    #A Labiodentals â€” keep LL & IM
    "/A": [("corners", {}), ("lip_band", {"which": "upper"})],

    #B Rounded-back vowels & rhotics/glide â€” keep PR
    "/B": [("corners", {}), ("inside_mouth", {})],

    #C Bilabials â€” keep UL+LL contact
    "/C": [("corners", {}), ("inside_mouth", {})],

    #D Diphthongs (/aw/) â€” keep JD (extensive), PR (medium)
    "/D": [("corners", {})],

    #E Interdentals â€” keep UL+LL parted, occasional tongue; small AP
    "/E": [("corners", {})],

    #F Postalveolar sibilants & affricates â€” keep PR funnel, small AP (IM often occluded)
    "/F": [("corners", {}), ("inside_mouth", {})],

    #G Mid-back rounded â€” keep subtle PR/LC, JD mid
    "/G": [("inside_mouth", {})],

    #H Alveolar sibilants â€” keep LC stretch, narrow AP, often teeth visible
    "/H": [
        ("corners", {"which": "both","pad":6, "ksize":45}),
    ],

    #I Unrounded/front-central (spread/neutral) â€” keep LC spread & AP
    "/I": [
        ("inside_mouth", {"pad": 12, "ksize": 81}),
        ("quad", {"idxs": (50, 52, 58, 56), "pad": 10, "ksize": 75}),
    ],

    # /J Alveolar non-sibilants â€” keep slight UL/LL parting, small JD; occasional tongue tip
    "/J": [("corners", {})],

    # /K Velars â€” visible lip cues weak; emphasise JD by de-emphasising lips
    "/K": [
        ("lip_band", {"which": "upper"}),
        ("lip_band", {"which": "lower"}),
        ("inside_mouth", {}),
    ],

    # /S Silence & closures / non-speech â€” neutral/variable
    "/S": [],
}
```

### call blur functions 

```python
def render_smart_blur_blobs_only(
    video_path: Path,
    pkl_path: Path,
    textgrid_path: Path,
    target_visemes,          # str or List[str]
    span_mode: str,          # "word" or "phone"
    out_path: Path,
    # default strengths per blob type
    mouth_pad=10, mouth_ksize=75,
    band_pad=10,  band_ksize=75,
    corner_pad=10, corner_ksize=75,
):
    """
    Blobs-only smart blur:
      - For frames where any *target viseme* is active, apply that viseme's blob list.
      - `target_visemes` can be a single string ("/A") or a list (["/E","/F",...]).
      - `span_mode` selects whether trigger spans are computed at word-level or phone-level.
    """

    # ---- load landmarks ----
    with open(pkl_path, "rb") as f:
        lms = pickle.load(f)

    # ---- open video ----
    cap = cv2.VideoCapture(str(video_path))
    if not cap.isOpened():
        raise RuntimeError(f"Could not open video: {video_path}")

    fps    = cap.get(cv2.CAP_PROP_FPS)
    width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    N      = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    if len(lms) != N:
        print(f"[WARN] landmarks length ({len(lms)}) != video frames ({N}). Using min length.")
        N = min(len(lms), N)

    # ---- normalize targets + build priority (used if spans overlap) ----
    if isinstance(target_visemes, str):
        target_visemes = [target_visemes]
    assert span_mode in ("word", "phone")
    priority = {v: i for i, v in enumerate(target_visemes)}

    # ---- compute spans (sec) tagged with which viseme to use ----
    spans_sec   = get_target_spans(textgrid_path, target_visemes, span_mode=span_mode, lead=LEAD_SEC, lag=LAG_SEC)
    spans_frame = spans_to_frames_viseme(spans_sec, fps=fps)

    # ---- per-frame viseme label ("" = no blur) ----
    viseme_at = np.array([""] * N, dtype=object)

    for v, f0, f1 in spans_frame:
        f0c = max(0, f0)
        f1c = min(N - 1, f1)
        if f1c < f0c:
            continue
        # Fill this interval with viseme v, breaking ties by priority order
        for f in range(f0c, f1c + 1):
            cur = viseme_at[f]
            if cur == "" or priority.get(v, 10**9) < priority.get(cur, 10**9):
                viseme_at[f] = v

    # ---- warn if any target viseme has no blob mapping ----
    missing = [v for v in target_visemes if not VISEME_TO_BLOBS.get(v)]
    if missing:
        print(f"[WARN] No blob mapping for: {missing} (those visemes will effectively do nothing)")

    # ---- video writer ----
    fourcc = cv2.VideoWriter_fourcc(*"mp4v")
    out_v  = cv2.VideoWriter(str(out_path), fourcc, fps, (width, height))

    for f in range(N):
        ok, frame = cap.read()
        if not ok:
            break
        lm = lms[f]

        v = viseme_at[f]
        blobs = VISEME_TO_BLOBS.get(v, []) if v else []

        if blobs:
            for kind, kwargs in blobs:
                if kind == "inside_mouth":
                    frame = blur_inside_mouth_blob(
                        frame, lm,
                        pad=int(kwargs.get("pad", mouth_pad)),
                        ksize=int(kwargs.get("ksize", mouth_ksize)),
                    )
                elif kind == "lip_band":
                    frame = blur_lip_band_blob(
                        frame, lm,
                        which=str(kwargs.get("which", "upper")),
                        pad=int(kwargs.get("pad", band_pad)),
                        ksize=int(kwargs.get("ksize", band_ksize)),
                    )
                elif kind == "corners":
                    frame = blur_corners_blob(
                        frame, lm,
                        which=str(kwargs.get("which", "both")),
                        pad=int(kwargs.get("pad", corner_pad)),
                        ksize=int(kwargs.get("ksize", corner_ksize)),
                    )
                elif kind == "quad" and "idxs" in kwargs:
                    frame = blur_specific_quad(
                        frame, lm,
                        idxs=tuple(kwargs["idxs"]),
                        pad=int(kwargs.get("pad", 10)),
                        ksize=int(kwargs.get("ksize", 75)),
                    )
                # unknown kinds are ignored

        out_v.write(frame)

    cap.release()
    out_v.release()
    print(f"Saved: {out_path}")
```

```python
# --- Cell 10: Batch runner for one speaker & one-or-many visemes (self-contained) ---
from typing import List, Tuple
from pathlib import Path

# --- Helpers (inlined so this cell works even after a kernel restart) ---

def viseme_folder_name(target_visemes, span_mode: str) -> str:
    """
    Examples:
      "/I" -> "Viseme_I_word"
      ["/E","/F","/H","/J","/K"] -> "Visemes_E_F_H_J_K_phone"
    """
    if isinstance(target_visemes, str):
        target_visemes = [target_visemes]
    codes = "_".join(v.strip("/").upper() for v in target_visemes)
    prefix = "Viseme" if len(target_visemes) == 1 else "Visemes"
    return f"{prefix}_{codes}_{span_mode}"

def out_dir_for_speaker(speaker_dir: Path, target_visemes, span_mode: str) -> Path:
    """
    <speaker>/Clips/processed/smart_blur/<Viseme(s)_...>  (separates word vs phone runs)
    """
    return speaker_dir / "Clips" / "processed" / "smart_blur" / viseme_folder_name(target_visemes, span_mode)

def find_speaker_dir(data_root: Path, group: str, hint: str) -> Path:
    """
    Robust resolver that accepts:
      - corpus root (contains 'volunteers'/'lipspeakers')
      - group dir itself (e.g. .../volunteers)
      - speaker dir (e.g. .../volunteers/21M)
      - a subfolder inside speaker dir (e.g. .../volunteers/21M/Clips or .../processed)
    Returns the speaker directory (e.g. .../volunteers/21M).
    """
    dr = Path(data_root)
    group_low = group.lower()
    hint_up = hint.upper()

    # --- Case 0: data_root is inside a speaker dir (or is speaker dir) ---
    # climb up a few levels looking for a directory that:
    #   - its name contains the hint (21 / 21M / etc)
    #   - AND has a "Clips" folder inside it
    for cand in [dr, dr.parent, dr.parent.parent, dr.parent.parent.parent]:
        try:
            if cand.exists() and cand.is_dir():
                if (cand / "Clips").exists() and (hint_up in cand.name.upper()):
                    return cand
        except Exception:
            pass

    # --- Determine the group directory (base) ---
    # Case A: dr is corpus root (dr/group exists)
    if (dr / group).exists():
        base = dr / group

    # Case B: dr is already the group directory
    elif dr.exists() and dr.is_dir() and dr.name.lower() == group_low:
        base = dr

    else:
        # Case C: dr is somewhere deeper; walk upwards to find a parent that contains the group dir
        base = None
        for p in [dr] + list(dr.parents):
            if (p / group).exists():
                base = p / group
                break

        if base is None:
            # helpful debug info
            tried = [
                f"{dr / group}",
                f"{dr} (as group?)",
                "parents containing <parent>/<group>"
            ]
            raise FileNotFoundError(
                "Could not locate group folder.\n"
                f"  data_root={dr}\n"
                f"  group={group}\n"
                f"  tried: {tried}\n"
                "Fix: set DATA_ROOT to corpus root (the folder that contains 'volunteers' and 'lipspeakers'),\n"
                "or set DATA_ROOT to the group folder directly (e.g. .../volunteers)."
            )

    # --- Now select the speaker dir inside base ---
    cands = [p for p in base.iterdir() if p.is_dir() and (hint_up in p.name.upper())]
    if not cands:
        raise FileNotFoundError(f"No speaker matches hint '{hint}' in {base}")

    if len(cands) > 1:
        names = ", ".join(p.name for p in cands)
        raise RuntimeError(f"Ambiguous speaker hint '{hint}'. Candidates: {names}")

    return cands[0]


def collect_clip_triplets(speaker_dir: Path, camera: str, video_subdir: str
                         ) -> List[Tuple[Path, Path, Path, str]]:
    """
    Return [(video, pkl, textgrid, code)] for all clips that have all three artifacts.
    Assumes structure:
      Clips/processed/{video_subdir}/{camera}/*.mp4
      Clips/processed/landmarks/{camera}/*.pkl
      Clips/aligned/{camera}/*.TextGrid
    """
    vdir = speaker_dir / "Clips" / "processed" / video_subdir / camera
    ldir = speaker_dir / "Clips" / "processed" / "landmarks" / camera
    tdir = speaker_dir / "Clips" / "aligned"   / camera

    if not vdir.exists(): raise FileNotFoundError(f"Missing video dir: {vdir}")
    if not ldir.exists(): raise FileNotFoundError(f"Missing landmarks dir: {ldir}")
    if not tdir.exists(): raise FileNotFoundError(f"Missing TextGrid dir: {tdir}")

    def by_stem(d: Path, ext: str):
        return {p.stem: p for p in d.glob(f"*.{ext}")}

    vids = by_stem(vdir, "mp4")
    pkl  = by_stem(ldir, "pkl")
    tgs  = by_stem(tdir, "TextGrid")

    codes = sorted(set(vids) & set(pkl) & set(tgs))
    triplets = [(vids[c], pkl[c], tgs[c], c) for c in codes]
    return triplets

def summarize_batch(triplets: List[Tuple[Path, Path, Path, str]], max_show=5):
    print(f"Found {len(triplets)} complete clip triplets.")
    for (v, p, t, c) in triplets[:max_show]:
        print(f"  {c}:")
        print(f"    video: {v.name}")
        print(f"    pkl  : {p.name}")
        print(f"    tg   : {t.name}")
    if len(triplets) > max_show:
        print(f"  ... ({len(triplets)-max_show} more)")

def process_speaker(
    data_root: Path,
    group: str,
    speaker_hint: str,
    target_visemes,          # str or List[str]
    span_mode: str,          # "word" or "phone"
    camera: str,
    video_subdir: str,
    mouth_pad=10, mouth_ksize=75,
    band_pad=10,  band_ksize=75,
    corner_pad=10, corner_ksize=75,
) -> None:
    spk_dir = find_speaker_dir(data_root, group, speaker_hint)
    triplets = collect_clip_triplets(spk_dir, camera, video_subdir)
    summarize_batch(triplets)

    out_root = out_dir_for_speaker(spk_dir, target_visemes, span_mode)
    out_root.mkdir(parents=True, exist_ok=True)

    for (video_path, pkl_path, textgrid_path, code) in triplets:
        out_path = out_root / video_path.name  # keep exact input filename
        
        # Skip if output already exists
        if out_path.exists():
            print(f"[{spk_dir.name}] {code} â†’ SKIPPED (already exists)")
            continue
        
        if WRITE_VIDEO:
            print(f"[{spk_dir.name}] {code} â†’ {out_path}")
            render_smart_blur_blobs_only(
                video_path=video_path,
                pkl_path=pkl_path,
                textgrid_path=textgrid_path,
                target_visemes=target_visemes,
                span_mode=span_mode,
                out_path=out_path,
                mouth_pad=mouth_pad, mouth_ksize=mouth_ksize,
                band_pad=band_pad,   band_ksize=band_ksize,
                corner_pad=corner_pad, corner_ksize=corner_ksize,
            )

for speaker_code in SPEAKERS_TO_PROCESS:
    print(f"{'='*60}\n")
    print(f"Processing {SPEAKER_GROUP_NAME}:", speaker_code)
    print(f"{'='*60}\n")
    # ---- Kick off the batch using your choices from Cell 1 ----
    process_speaker(
        data_root=DATA_ROOT,
        group=SPEAKER_GROUP_NAME,
        speaker_hint=speaker_code,
        target_visemes=TARGET_VISEMES,
        span_mode=SPAN_MODE,
        camera=CAMERA,
        video_subdir=VIDEO_SUBDIR,
)
```

### Viseme Family Visual Grid (Thesis Figure)
Generate a composite visual showing one representative frame per viseme family (/A through /K, plus /S) with the corresponding blur applied. This creates a single-page figure for your thesis showing the blurring effect for each viseme group.

**Process:**
1. Scan all 01M clips to find the longest (most visible) span for each viseme
2. Extract the middle frame of that span
3. Apply the blur style defined in `VISEME_TO_BLOBS` for that viseme
4. Arrange into a 3Ã—4 grid with viseme labels and descriptions
5. Save as high-quality PNG for thesis inclusion

```python
# Build viseme grid for thesis figure
from pathlib import Path
from typing import Dict, Tuple, Optional
import numpy as np
import cv2
import math

# Config for viseme grid generation
THESIS_GRID_SPEAKER = "01M"
THESIS_GRID_CAMERA = "straightcam"
THESIS_GRID_VIDEO_SUBDIR = "video25"
THESIS_GRID_OUT_DIR = DATA_ROOT / "thesis_viseme_grids"
THESIS_GRID_OUT_DIR.mkdir(parents=True, exist_ok=True)

# Viseme metadata for grid labels
VISEME_LABELS = {
    "/A": ("Labiodentals (f, v)", (0, 0, 255)),
    "/B": ("Rounded-back vowels & rhotics (u, É¹, w)", (255, 0, 0)),
    "/C": ("Bilabials (b, p, m)", (0, 255, 0)),
    "/D": ("Diphthongs (aÊŠ)", (0, 255, 255)),
    "/E": ("Interdentals (Ã°, Î¸)", (255, 0, 255)),
    "/F": ("Postalveolars (Êƒ, tÊƒ)", (255, 255, 0)),
    "/G": ("Mid-back rounded (É”Éª)", (128, 128, 255)),
    "/H": ("Alveolar sibilants (s, z)", (255, 128, 0)),
    "/I": ("Unrounded/front spread (a, e, i)", (128, 255, 128)),
    "/J": ("Alveolar non-sibilants (t, d, n, l)", (255, 128, 255)),
    "/K": ("Velars (g, k, Å‹)", (128, 255, 255)),
    "/S": ("Silence/Closures (h, Ê”)", (200, 200, 200)),
}

ALL_VISEMES = list(VISEME_LABELS.keys())

def find_longest_viseme_span(triplets, viseme: str, fps: float) -> Optional[Tuple[Path, Path, Path, int]]:
    """
    Scan through all triplets, find the longest span for the given viseme.
    Return (video_path, pkl_path, textgrid_path, frame_idx) for the middle of the longest span.
    """
    best_span = None
    best_dur = 0
    best_triplet = None
    
    for (video_path, pkl_path, tg_path, code) in triplets:
        try:
            spans_sec = get_target_spans(tg_path, [viseme], span_mode="phone", lead=LEAD_SEC, lag=LAG_SEC)
            for v, t0, t1 in spans_sec:
                dur = t1 - t0
                if dur > best_dur:
                    best_dur = dur
                    best_span = (t0, t1)
                    best_triplet = (video_path, pkl_path, tg_path, code)
        except Exception as e:
            pass  # skip triplets with errors
    
    if best_span is None:
        return None
    
    t0, t1 = best_span
    mid_frame = int(round((t0 + t1) / 2 * fps))
    return best_triplet + (mid_frame,)

def extract_and_blur_frame(video_path: Path, pkl_path: Path, frame_idx: int, viseme: str) -> Optional[np.ndarray]:
    """
    Extract a single frame, apply the blur for the given viseme, add label, return the blurred frame.
    """
    try:
        # Load landmarks
        with open(pkl_path, "rb") as f:
            lms = pickle.load(f)
        
        # Open video and extract frame
        cap = cv2.VideoCapture(str(video_path))
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_idx)
        ret, frame = cap.read()
        cap.release()
        
        if not ret or frame is None:
            return None
        
        lm = lms[frame_idx]
        
        # Apply viseme-specific blurring
        blobs = VISEME_TO_BLOBS.get(viseme, [])
        for kind, kwargs in blobs:
            if kind == "inside_mouth":
                frame = blur_inside_mouth_blob(
                    frame, lm,
                    pad=int(kwargs.get("pad", 10)),
                    ksize=int(kwargs.get("ksize", 75)),
                )
            elif kind == "lip_band":
                frame = blur_lip_band_blob(
                    frame, lm,
                    which=str(kwargs.get("which", "upper")),
                    pad=int(kwargs.get("pad", 10)),
                    ksize=int(kwargs.get("ksize", 75)),
                )
            elif kind == "corners":
                frame = blur_corners_blob(
                    frame, lm,
                    which=str(kwargs.get("which", "both")),
                    pad=int(kwargs.get("pad", 10)),
                    ksize=int(kwargs.get("ksize", 75)),
                )
            elif kind == "quad" and "idxs" in kwargs:
                frame = blur_specific_quad(
                    frame, lm,
                    idxs=tuple(kwargs["idxs"]),
                    pad=int(kwargs.get("pad", 10)),
                    ksize=int(kwargs.get("ksize", 75)),
                )
        
        # Add label to frame
        label_text = f"{viseme} - {VISEME_LABELS[viseme][0]}"
        cv2.putText(frame, label_text, (10, 30),
                   cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2, cv2.LINE_AA)
        
        return frame
    except Exception as e:
        print(f"  [ERROR] {e}")
        return None

# --- Main execution ---
print(f"Building viseme frames for speaker {THESIS_GRID_SPEAKER}...")

try:
    spk_dir = find_speaker_dir(DATA_ROOT, SPEAKER_GROUP_NAME, THESIS_GRID_SPEAKER)
    print(f"Found speaker dir: {spk_dir}")
    
    triplets = collect_clip_triplets(spk_dir, THESIS_GRID_CAMERA, THESIS_GRID_VIDEO_SUBDIR)
    print(f"Found {len(triplets)} clips")
    
    # Get FPS from first triplet to convert times to frames
    cap = cv2.VideoCapture(str(triplets[0][0]))
    fps = cap.get(cv2.CAP_PROP_FPS)
    cap.release()
    print(f"FPS: {fps}")
    
    for viseme in ALL_VISEMES:
        print(f"\nLooking for viseme {viseme}...")
        result = find_longest_viseme_span(triplets, viseme, fps)
        
        if result is None:
            print(f"  [SKIP] No frames found")
        else:
            video_path, pkl_path, tg_path, code, frame_idx = result
            print(f"  Found longest span in clip: {code} at frame {frame_idx}")
            frame = extract_and_blur_frame(video_path, pkl_path, frame_idx, viseme)
            if frame is not None:
                out_path = THESIS_GRID_OUT_DIR / f"{THESIS_GRID_SPEAKER}_viseme_{viseme.strip('/')}.png"
                cv2.imwrite(str(out_path), frame)
                print(f"  âœ“ Saved: {out_path}")
            else:
                print(f"  [FAIL] Could not extract frame")
    
    print(f"\nâœ“ All frames saved to: {THESIS_GRID_OUT_DIR}")

except Exception as e:
    print(f"[ERROR] {e}")
    import traceback
    traceback.print_exc()
```
